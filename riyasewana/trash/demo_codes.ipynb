## Inference

```python
import pandas as pd
from rapidfuzz import process
import difflib
import spacy

model_name="CAMRY2000"

df=pd.read_csv('transformed_table_data.csv')
df['COMBINED'] = (df['model'] + ' ' + df['Trim / Edition']).str.strip()

print("----using rapidfuzz-----")
choices = df['COMBINED'].unique()
match = process.extractOne(model_name, choices)
print(match)


print("----using difflib-----")
matches = difflib.get_close_matches(model_name, df['COMBINED'].unique(), n=3, cutoff=0.6)
print(matches)


print("----using spacy model-----")
nlp = spacy.load("en_core_web_sm")
query = nlp(model_name)
texts = [f"{row['model']} {row['Trim / Edition']}" for _, row in df.iterrows()]
docs = list(nlp.pipe(texts))
similarities = [query.similarity(doc) for doc in docs]
matches = [(i, sim) for i, sim in enumerate(similarities) if sim > 0.8]
similar_rows = df.loc[[index for index, _ in matches]]
# Sort by highest similarity
matches = sorted(matches, key=lambda x: x[1], reverse=True)
# Get the index of the highest similarity match
top_index = matches[0][0]
top_record = df.loc[top_index]
combined_value = top_record['COMBINED']
print(combined_value)
```
